# Rozkład chi kwadrat z $n$ stopniami swobody
Rozkład chi kwadrat ma gęstość:

$$
f_{X}(x) = 
\left\{ \begin{array}{ll}
{\frac {1}{2^{\frac {n}{2}}\Gamma \left({\frac {n}{2}}\right)}}\;x^{{\frac {n}{2}}-1}e^{-{\frac {x}{2}}} 
   & \textrm{gdy } x \geq 0\\
0 & \textrm{gdy } x < 0 
\end{array} \right. ,
$$

gdzie $\Gamma$ oznacza funkcję Gamma https://pl.wikipedia.org/wiki/Funkcja_\%CE\%93


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

# Definiujemy zmienną losową

```python
n=4
chi2 = stats.chi2(n)
```

# Narysujmy gęstość zmiennej losowej

```python
x = np.linspace(0, 100, 100)
plt.plot(x, stats.chi2.pdf(x, 50))
plt.show()
```

# Narysujmy gęstości rozkładu z różnymi parametrami

# Narysujmy kilka histogramów różnych próbek z rozkładu o tych samych parametrach.

```python
def get_cmap(n, name='hsv'):
    return plt.cm.get_cmap(name, n)

fig, ax = plt.subplots(1,1)
x=np.arange(-10,100)
cmap=get_cmap(40)

for n in np.arange(1,50, 10):
    ax.plot(x,stats.chi2.pdf(x, n),'bo--', label=f'n={n}', color=cmap(n))
ax.legend()
```
